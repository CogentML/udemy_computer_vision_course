```python
def imshow(title="Image", image=None, size=5):
    if image is None:
        print("Error: Image is None.")
        return

    try:
        w, h, _ = image.shape  # Get the width and height of the image
        aspect_ratio = h / w  # Calculate the aspect ratio

        # Adjust the figure size based on the aspect ratio and specified size
        plt.figure(figsize=(size, size * aspect_ratio))

        # Convert BGR to RGB for display
        rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        plt.imshow(rgb_image)
        plt.title(title)
        plt.show()
    except cv2.error as e:
        print(f"Error displaying image: {e}")
```

```python
def imshow(title = "", image = None, size = 10):

    # The line below is changed from w, h to h, w
    h, w = image.shape[0], image.shape[1]
    aspect_ratio = w/h

    plt.figure(figsize=(size * aspect_ratio, size))
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title(title)
    plt.show()
```

```python
# Example usage:
image_path = os.path.join("data", "images", "Sunflowers.jpg")
image = cv2.imread(image_path)

# You can specify the size parameter (e.g., size=8) if needed
imshow("Sunflowers Image", image)
```
